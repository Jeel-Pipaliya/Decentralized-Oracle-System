```python
final=pd.DataFrame()
for a in range(1,199):
    url='https://internshala.com/internships/page-{}/'.format(a)
    webpage=requests.get(url).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_="internship_meta duration_meta")
    name=[]
    stipend=[]
    location=[]
    hiring=[]
    all_skills = [] 
    for i in company:
        name.append(i.find("p").text.strip())
        stipend.append(i.find("span",class_="stipend").text.removesuffix("/month"))
        location.append(i.find("a").text)
        badge = i.find("div", class_="actively-hiring-badge")
        hiring.append(badge.text if badge else "Not actively hiring")

        company_skills = []
        skill_divs = i.find_all("div", class_="job_skill")
        for div in skill_divs:
            company_skills.append(div.text.strip())
        all_skills.append(company_skills)
    df = pd.DataFrame({
    'Name':name,'location':location,'stipend':stipend,"company_id": range(len(all_skills)),
    "skills": all_skills,"hiring":hiring
    })
    final = pd.concat([final, df], ignore_index=True)
```

```python
final
```

```python
file_path ="data.csv"
```

```python
final.to_csv(file_path, index=False, encoding="utf-8")
```
